```python
import numpy as np
from sklearn.model_selection import train_test_split

X, y = np.arange(1000).reshape((100, 10)), np.arange(100)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)   # 70 : 30
print(X_train.shape, X_test.shape)   # (70,10) (30,10)
```

```python
import torch, torch.nn as nn, torch.optim as optim

model = nn.Linear(1, 1)
criterion = nn.MSELoss()                       # 손실 함수
optimizer = optim.SGD(model.parameters(), lr=0.01)

X = torch.tensor([[1.],[2.],[3.],[4.]])
y = torch.tensor([[2.],[4.],[6.],[8.]])        # y = 2x

for epoch in range(200):                       # 외부 반복(에포크)
    optimizer.zero_grad()    # (1) 기울기 초기화
    outputs = model(X)       # (2) 예측값 산출
    loss = criterion(outputs, y)  # (3) 손실 계산
    loss.backward()          # (4) 역전파
    optimizer.step()         # (5) 가중치 업데이트
print("예측:", model(torch.tensor([[5.]])).item())   # 10에 근접
```
