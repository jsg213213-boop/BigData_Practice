```python
import torch
x = torch.rand(3, 3)     # 3x3 무작위 텐서
y = torch.rand(3, 3)
print("x :",x)
print("y :",y)
z = x + y                # 텐서 덧셈
print(z)
print("shape:", z.shape) # 모양 확인
```

```python
import torch
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print(x)
print("shape:", x.shape)   # torch.Size([2, 3])
print("dtype:", x.dtype)
```

```python
w = torch.tensor(2.0, requires_grad=True)   # 기울기 추적 켜기
y = w ** 2
z = 2 * y + 5         # z = 2w^2 + 5
z.backward()          # dz/dw 자동 계산
print(w.grad)         # dz/dw = 4w = 4*2.0 = 8.0
```

```python
import torch.nn as nn
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(in_features=10, out_features=1)  # 입력 10 → 출력 1
    def forward(self, x):
        return self.linear(x)   # 순전파

model = MyModel()
print(model)
```
