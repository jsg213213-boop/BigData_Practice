```python
import torch, torch.nn as nn
x = torch.tensor([-2.0, -0.5, 0.0, 1.5, 3.0])
print("ReLU   :", nn.ReLU()(x))       # 음수는 0
print("Sigmoid:", nn.Sigmoid()(x))    # 0~1
print("Tanh   :", nn.Tanh()(x))       # -1~1
```

```python
w = torch.tensor(2.0, requires_grad=True)
loss = (w - 5) ** 2          # 최소가 되는 w는 5
loss.backward()              # 역전파
print(w.grad)                # d/dw (w-5)^2 = 2(w-5) = -6
```
