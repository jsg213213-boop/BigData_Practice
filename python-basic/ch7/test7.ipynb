```python
import pymysql

con = pymysql.connect(
    host='127.0.0.1',
    user='webuser',
    password='webuser',
    database='webdb',
    charset='utf8')
print('DB 연결 성공!')

cur = con.cursor()   # 커서 생성 (SQL 실행 통로)
```

```python
sql = """
CREATE TABLE book2 (
    id INT AUTO_INCREMENT PRIMARY KEY,
    title VARCHAR(200),
    pub VARCHAR(45),
    pages INT,
    author VARCHAR(100)
)
"""
cur.execute(sql)
print('테이블 생성 성공!')
```

```python
sql = "INSERT INTO book2 (title, pub, pages, author) VALUES (%s, %s, %s, %s)"
cur.execute(sql, ('머신러닝 입문', '길볏', 320, '박민수'))
con.commit()   # ← 없으면 저장 안 됨!
print('1개 행 추가됨')
```

```python
cur.execute('SELECT * FROM book2')
rows = cur.fetchall()        # 모든 행
for row in rows:
    print(row)
# (1, '파이썬 기본서', '한빛미디어', 300, '김철수')

row = cur.fetchone()         # 한 행만
rows2 = cur.fetchmany(2)     # n개만 (페이지네이션)
```

```python
# 수정
cur.execute('UPDATE book2 SET pages = %s WHERE id = %s', (350, 1))
con.commit()

# 삭제
cur.execute('DELETE FROM book2 WHERE id = %s', (1,))
con.commit()
```

```python
# DictCursor: 값을 이름으로 꺼내기
cur = con.cursor(pymysql.cursors.DictCursor)
cur.execute('SELECT * FROM book2 WHERE pages > %s ORDER BY pages DESC LIMIT 5', (200,))
for row in cur.fetchall():
    print(row['title'], row['pages'])

# Pandas와 연결: SQL 결과를 바로 DataFrame으로
import pandas as pd
df = pd.read_sql('SELECT * FROM book2', con)
print(df.head())
```

```python
# 1. 혹시 기존에 잘못 만들어진 product 테이블이 있다면 삭제
cur.execute("DROP TABLE IF EXISTS product")

# 2. name과 price 컬럼을 모두 포함해서 테이블 새로 정의
sql = """
CREATE TABLE product (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(100),
    price INT
)
"""
cur.execute(sql)
print("1단계: product 테이블 재생성 완벽 성공!")
```

```python
sql = "INSERT INTO product (name, price) VALUES (%s, %s)"
cur.execute(sql, ('노트북', 1200000))
con.commit()
print('삽입 완료')
```

```python
cur.execute('SELECT * FROM product WHERE price >= %s', (1000000,))
for row in cur.fetchall():
    print(row)
```
