```python
x = input("숫자를 입력하세요: ")  # input은 항상 '문자열'로 받음
x = int(x)                        # 정수로 형변환
print("제곱:", x * x)
```

```python
score = 85
if score >= 90:
    print("A")
elif score >= 80:
    print("B")   # 이 줄이 실행됨
else:
    print("C 이하")
```

```python
# 1부터 5까지 합
total = 0
for i in range(1, 6):   # 1,2,3,4,5
    total += i
print("합:", total)     # 합: 15
```

```python
name = "정성규"      # 문자열(str)
age = 33             # 정수(int)
height = 160.5       # 실수(float)
is_student = True    # 불리언(bool)

print(type(name), type(age), type(height), type(is_student))
```

```python
count = 3
while count > 0:
    print("카운트다운:", count)
    count -= 1
print("발사!")
```

```python
for i in range(1, 11):
    if i % 2 == 0:   # 2로 나눈 나머지가 0 = 짝수
        print(i)
# 또는 한 줄로: for i in range(2, 11, 2): print(i)
```
