```python
def add(a, b):       # a, b는 매개변수
    return a + b     # 결과 반환

result = add(1, 2)   # 1, 2는 인수
print(result)        # 3
```

```python
def greet(name, msg="안녕하세요"):
    print(f"{name}님, {msg}")

greet("철수")              # 철수님, 안녕하세요
greet("영희", "반갑습니다")  # 영희님, 반갑습니다
```

```python
def profile(**info):   # info는 딕셔너리로 묶임
    for key, value in info.items():
        print(f"{key}: {value}")

profile(이름="정성규", 직업="수강생", 지역="부산")
```

```python
def change(items):
    items.append(99)    # 원본 리스트가 바뀜! (가변)

my_list = [1, 2, 3]
change(my_list)
print(my_list)          # [1, 2, 3, 99]
```

```python
# 문자열 슬라이싱 (ch3 예제)
s = "abc"
print(s[0:2])   # ab  (인덱스 0~1)
print(s[::-1])  # cba (거꿁로)

# lambda 예시
double = lambda x: x * 9
print(double(7))  # 63
```

```python
def area(r):
    return 3.14 * r * r
print(area(5))   # 78.5
```

```python
def power(base, exp=2):
    return base ** exp
print(power(3))      # 9
print(power(2, 3))   # 8
```

```python
def average(*nums):
    if len(nums) == 0:
        return 0
    return sum(nums) / len(nums)
print(average(80, 90, 100))  # 90.0
```

```python
def is_palindrome(s):
    return s == s[::-1]
print(is_palindrome("level"))  # True
print(is_palindrome("python")) # False
```
