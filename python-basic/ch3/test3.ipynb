```python
fruits = ["사과", "바나나", "포도"]
print(fruits[0])      # 사과 (첫 번째)
fruits.append("귀")  # 추가
fruits[1] = "딸기"   # 수정
print(fruits)         # ['사과', '딸기', '포도', '귀']
print(len(fruits))    # 4 (개수)
```

```python
point = (37.5, 127.0)   # 위도, 경도 (바뀌면 안 되는 값)
print(point[0])         # 37.5
# point[0] = 0  → 에러! 튜플은 수정 불가
```

```python
student = {"이름": "정성규", "나이": 33, "점수": 93}
print(student["이름"])       # 정성규
student["지역"] = "부산"     # 추가
for key, value in student.items():
    print(key, "=", value)
```

```python
nums = [1, 2, 2, 3, 3, 3]
unique = set(nums)       # {1, 2, 3}
print(unique)
print(len(unique))       # 3
```

```python
# 리스트 컴프리헨션 — 한 줄로 리스트 만들기 (빅데이터 전처리에 매우 중요)
squares = [x**2 for x in range(1, 6)]
print(squares)   # [1, 4, 9, 16, 25]

# 조건 붙이기
evens = [x for x in range(10) if x % 2 == 0]
print(evens)     # [0, 2, 4, 6, 8]
```

```python
scores = [70, 85, 90, 60, 95]
print("최대:", max(scores))               # 95
print("최소:", min(scores))               # 60
print("평균:", sum(scores) / len(scores)) # 80.0
```

```python
prices = {"사과":1000, "바나나":1500}
prices["포도"] = 2000
print(sum(prices.values()))   # 4500
```
