```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = ["영화 정말 재미있다", "정말 지루한 영화", "배우 연기 최고"]
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(docs)        # 문서 × 단어 행렬
print(tfidf.get_feature_names_out()) # 단어 목록
print(X.toarray().round(2))          # TF-IDF 값
```
